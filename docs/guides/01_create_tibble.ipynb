# <span style="color:#1E90FF">1. Create dataframe using datar.tibble.tibble()</span>

```python
tb = dr.tibble(
    x = [1, 2, 3],
    y = ["a", "b", "c"],
    z = [True, False, True]
)

print(tb)
```

# <span style="color:#1E90FF">2. Convert polars DataFrame to datar tibble</span>

```python
df = pl.DataFrame({
    'A': [1, 2, 3],
    'B': ['x', 'y', 'z'],
    'C': [True, False, True]
})

tb_from_df = dr.tibble(df)

print(tb_from_df)
```
