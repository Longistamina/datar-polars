```python
1. dr.slice_head() - Return the first n rows.

2. dr.slice_tail() - Return the last n rows.

3. dr.glimpse() - Transpose print of tibble, showing data types and a preview of data (like df.info())
```

````python
import datar.all as dr
from datar import f

import polars as pl

from pathlib import Path
import warnings

warnings.filterwarnings("ignore")
pl.Config.set_tbl_cell_alignment("CENTER")

data_dir = next(Path("/home/").glob("**/datar-polars/data"))
````

````python
tb_pokemon = (
    dr.tibble(pl.read_csv(data_dir/'pokemon.csv'))
    >> dr.select(~f['#'])
    >> dr.rename_with(lambda col: pl.Series([col]).str.replace(r"\s+", "_").str.replace(".", "", literal=True).item())
    >> dr.pipe(lambda f: f >> dr.mutate(
        Type_1 = f.Type_1.cast(pl.Categorical),
        Type_2 = f.Type_2.cast(pl.Categorical),
        Generation = f.Generation.cast(pl.String).cast(
            pl.Enum(f.Generation.cast(pl.String).unique().sort().to_list())
        ),
        Legendary = dr.as_logical(f.Legendary)
    ))
)

print(tb_pokemon)
````

# <span style="color:#1E90FF">1. dr.slice_head()</span>

````python
print(
    tb_pokemon >> dr.slice_head(n=3) # Show the first 3 rows
)
````

````python
print(
    tb_pokemon
    >> dr.slice_head() # Show the first 1 row (default)
)
````

# <span style="color:#1E90FF">2. dr.slice_tail()</span>

````python
print(
    tb_pokemon >> dr.slice_tail(n=3) # Show the last 3 rows
)
````

````python
print(
    tb_pokemon
    >> dr.slice_tail() # Show the last 1 rows (default)
)
````

# <span style="color:#1E90FF">3. dr.glimpse()</span>

## NOTE: dr.glimpse(tb_pokemon) DOES NOT WORK, must use the pipe operator (>>)

````python
tb_pokemon >> dr.glimpse()
````
